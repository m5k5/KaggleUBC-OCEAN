```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import seaborn as sns

sns.set_theme()


DATA_PATH = "E:/ML/UBC"
print(DATA_PATH)
```

```python
df1 = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "ResNet2DModelThumbnails_K_4_Acc.csv"))
acc = df1["Value"].to_numpy()
accSteps = df1["Step"].to_numpy()

df2 = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "ResNet2DModelThumbnails_K_4_ValAcc.csv"))
val = df2["Value"].to_numpy()
valSteps = df2["Step"].to_numpy()
```

```python
from scipy.ndimage import gaussian_filter1d

plt.rc('font', size=14) 
plt.rc('axes', titlesize=16, labelsize=16)     # fontsize of the axes title
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels
plt.rc('legend', fontsize=16)    # legend fontsize
plt.rc('figure', titlesize=20)

plt.figure(figsize=(10,14))

plt.subplot(2,1,1)
# _ = plt.plot(gaussian_filter1d(acc, sigma=3), label="Training Accuracy")
sns.lineplot(x=accSteps, y=gaussian_filter1d(acc, sigma=3), label="Training Accuracy")
_ = plt.xlabel("Steps")
_ = plt.ylabel("Accuracy")

plt.subplot(2,1,2)
_ = sns.lineplot(x=valSteps, y=gaussian_filter1d(val, sigma=3), label="Validation Accuracy")
_ = plt.xlabel("Epochen")
_ = plt.ylabel("Accuracy")

_ = plt.legend()

_ = plt.suptitle("Thumbnail Model")

# plt.tight_layout()
```

```python
from scipy.ndimage import gaussian_filter1d

plt.rc('font', size=14) 
plt.rc('axes', titlesize=16, labelsize=16)     # fontsize of the axes title
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels
plt.rc('legend', fontsize=16)    # legend fontsize
plt.rc('figure', titlesize=20)

plt.figure(figsize=(10,14))

plt.subplot(2,1,1)
# _ = plt.plot(gaussian_filter1d(acc, sigma=3), label="Training Accuracy")
sns.lineplot(x=accSteps, y=gaussian_filter1d(acc, sigma=8), label="Training Accuracy")
_ = plt.xlabel("Steps")
_ = plt.ylabel("Accuracy")

plt.subplot(2,1,2)
_ = sns.lineplot(x=valSteps, y=gaussian_filter1d(val, sigma=3), label="Validation Accuracy")
_ = plt.xlabel("Epochen")
_ = plt.ylabel("Accuracy")

_ = plt.legend()

_ = plt.suptitle("Patches Model")

# plt.tight_layout()
```

```python
df1 = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "edgenextBase_384_Acc.csv"))
accEdge = df1["Value"].to_numpy()
accStepsEdge = df1["Step"].to_numpy()

df2 = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "edgenextBase_384_ValAcc.csv"))
valEdge = df2["Value"].to_numpy()
valStepsEdge = df2["Step"].to_numpy()

df1 = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "resnet34d_384_Acc.csv"))
accRN34d = df1["Value"].to_numpy()
accStepsRN34d = df1["Step"].to_numpy()

df2 = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "resnet34d_384_ValAcc.csv"))
valRN34d = df2["Value"].to_numpy()
valStepsRN34d = df2["Step"].to_numpy()

df1 = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "tf_efficientnet_b4_Acc.csv"))
accTfEffNet = df1["Value"].to_numpy()
accStepsTfEffNet = df1["Step"].to_numpy()

df2 = pd.read_csv(os.path.join(DATA_PATH, "trainingHistory", "tf_efficientnet_b4_ValAcc.csv"))
valTfEffNet = df2["Value"].to_numpy()
valStepsTfEffNet = df2["Step"].to_numpy()
```

```python
from scipy.ndimage import gaussian_filter1d

plt.rc('font', size=14) 
plt.rc('axes', titlesize=16, labelsize=16)     # fontsize of the axes title
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels
plt.rc('legend', fontsize=16)    # legend fontsize
plt.rc('figure', titlesize=20)

plt.figure(figsize=(10,14))

plt.subplot(2,1,1)
# _ = plt.plot(gaussian_filter1d(acc, sigma=3), label="Training Accuracy")
plt.title("Training")
sns.lineplot(x=accStepsEdge, y=gaussian_filter1d(accEdge, sigma=8), label="EdgeNext Base")
sns.lineplot(x=accStepsRN34d, y=gaussian_filter1d(accRN34d, sigma=8), label="ResNet34d")
sns.lineplot(x=accStepsTfEffNet, y=gaussian_filter1d(accTfEffNet, sigma=8), label="TF EfficientNet B4")
_ = plt.xlabel("Steps")
_ = plt.ylabel("Accuracy")

plt.subplot(2,1,2)
plt.title("Validation")
sns.lineplot(x=valStepsEdge, y=gaussian_filter1d(valEdge, sigma=1), label="EdgeNext Base")
sns.lineplot(x=valStepsRN34d, y=gaussian_filter1d(valRN34d, sigma=1), label="ResNet34d")
sns.lineplot(x=valStepsTfEffNet, y=gaussian_filter1d(valTfEffNet, sigma=1), label="TF EfficientNet B4")
_ = plt.xlabel("Epochen")
_ = plt.ylabel("Accuracy")

_ = plt.legend()

_ = plt.suptitle("Patches Model")

# plt.tight_layout()
```

```python
import numpy as np
from PIL import Image
import itertools as it
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
import pandas as pd

DATA_PATH="E:/ML/UBC/train_images_test"
```

```python
pixelRange = np.arange(1000,41000,8000)
pixelRange
```

```python
table = []
for i, (w, h) in tqdm(enumerate(it.product(pixelRange, repeat=2)), total=pixelRange.shape[0]**2):
    im = np.ones((w, h, 3))*255
    im = Image.fromarray(im.astype(np.uint8))
    im.save(os.path.join(DATA_PATH, str(i)+".png"), optimize=True, quality=20)
    table.append([str(i), str(w), str(h)])
```

```python
df = pd.DataFrame(table, columns=["image_id","image_width", "image_height"])
df.to_csv(os.path.join(DATA_PATH, "test.csv"), index=False)
```
